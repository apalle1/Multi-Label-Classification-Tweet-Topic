# Extract tweets from Twitter API

API stands for Application Programming Interface. It is a tool that makes interaction with web services easy. Many web services provide APIs to developers to interact with their services and to access data programmatically. To access Twitter API, we need to get 4 pieces of information from Twitter: API key, API secret, Access token, and Access token secret. If you have no idea what these twitter credentials are, you must first become open up a twitter developer account and I’m sure you will find plenty of tutorials here in Medium and also youtube got heaps of videos on how to do this.

Pass the consumer_key & consumer_secret credentials to Tweepy’s OAuthHandler instance named ‘auth’, then using that instance call the method set_access_token by passing the access_key and, access_secret. After that we are setting up an instance called api. We will be using this api object to basically talk to twitter (to read data from twitter).

```
def limitHandled(cursor):
    while True:
        try:
            yield cursor.next()
        except tweepy.TweepError:
            time.sleep(15 * 60)

topics = ['politics','politics and travel','politics and car','electronics and car','electronics','car','travel','travel and electronics','travel and car']
if __name__ == '__main__':
    for topic in topics:
        users = {}
        for tweet in limitHandled(tweepy.Cursor(api.search, q=(topic), count=100, lang='en').items(5000)):
            print(tweet.user.name)
            print(tweet.text)
            if tweet.user.id in users:
                users[tweet.user.id].append(tweet.text)  
            else:
                users[tweet.user.id] = []
                users[tweet.user.id].append(tweet.text)
        fileName = topic + '_train_data.data'
        print (fileName)
        userTweets = open(fileName,'ab+')
        pickle.dump(users,userTweets)
        userTweets.close()
```
